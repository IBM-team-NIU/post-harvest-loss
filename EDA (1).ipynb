```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

file_path = 'merged_data.csv'
data = pd.read_csv(file_path)

label_encoder = LabelEncoder()

data['State_Name'] = label_encoder.fit_transform(data['State_Name'])

data['Crop'] = label_encoder.fit_transform(data['Crop'])
data['District_Name'] = label_encoder.fit_transform(data['District_Name'])



print(data.head())
```

```python
print("Missing values in merged DataFrame:\n",data.isnull().sum())
```

```python
data.to_csv('categorial.csv', index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'categorial.csv'
data = pd.read_csv(file_path)
print("Data Information:")
print(data.info())
print("\nSummary Statistics:")
print(data.describe())
print("\nMissing Values:")
print(data.isnull().sum())

numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(20, 15))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(5, 5, i)
    sns.histplot(data[column].dropna(), kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()

categorical_columns = data.select_dtypes(include=['object']).columns

plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(5, 5, i)
    sns.countplot(y=data[column])
    plt.title(column)
plt.tight_layout()
plt.show()

corr_matrix = data.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

plt.figure(figsize=(15, 10))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
plt.savefig('sample_plot.png', format='png')

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'categorial.csv'
data = pd.read_csv(file_path)

print("Data Information:")
print(data.info())
print("\nSummary Statistics:")
print(data.describe())
print("\nMissing Values:")
print(data.isnull().sum())

numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(20, 15))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(5, 5, i)
    sns.histplot(data[column].dropna(), kde=True)
    plt.title(column)
plt.tight_layout()
plt.savefig('numerical_distributions.png', format='png')  # Save as PNG
plt.show()  

categorical_columns = data.select_dtypes(include=['object']).columns

plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(5, 5, i)
    sns.countplot(y=data[column])
    plt.title(column)
plt.tight_layout()
plt.savefig('categorical_distributions.png', format='png') 
plt.show()  

corr_matrix = data.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.savefig('correlation_matrix.png', format='png')  
plt.show()  

plt.figure(figsize=(15, 10))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.savefig('missing_values_heatmap.png', format='png')  
plt.show()  
```
